```python
%matplotlib inline
import plot_helpers as ph
```

```python
graphs = [('DIbinary', 'accuracy')]
adult_datafiles = [
    ('../results/parity_tar/adult_race_numerical-binsensitive.csv', 'Adult (race)'),
    ('../results/parity_tar/adult_sex_numerical-binsensitive.csv', 'Adult (sex)'),
]
```

```python
for xaxis, yaxis in graphs:
    ph.plot_all(adult_datafiles, xaxis, yaxis, mapping=ph.mark_as_unfilled('UGP_dem_par_in_True'))
```

```python
def label_change(label):
    num = label.split('_')[-1]
    return f"target rate = {num}", False

data = ph.parse_all(adult_datafiles, ph.start_filter('UGP_dem_par_in_True'), label_change)
ph.generate_graph(data, ('DIbinary', '$DI$'), 'accuracy')
```

```python
adult_race = ph.parse('../results/parity_tar/adult_race_numerical-binsensitive.csv')
def real_tprdiff(fake_tprdiff):
    return 1 - fake_tprdiff
adult_race_t = ph.transform(adult_race, 'race-TPRDiff', real_tprdiff)
ph.generate_graph([ph.PlotDef('TPR difference (race)', adult_race_t)], ('race-TPRDiff', 'TPR diff'), 'accuracy')
```
